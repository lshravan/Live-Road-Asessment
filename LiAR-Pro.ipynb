```python
# get all columns
# used function Random to get total no.of columns and rows
import random
columns = []
for column_name in data.columns:
  # print (column_name, end="\t")
  columns.append(column_name)
print ("\nlength =", len(data.columns))
# number of rows in dataset
the_chosen_one = random.choice(columns)
number_of_rows = len(data[the_chosen_one])
print (number_of_rows)
```

```python
# divide the dataset into some limit size
# limit_size = 1000
LIMIT = 10000
iterations = min(LIMIT, number_of_rows)
new_data= [columns[:]]
for index in range(iterations):
  row = []
  nans = 0
  for column_name in columns:
    row.append(data[column_name][index])
    if str(data[column_name][index]) == "nan":
      nans += 1
  if len(row) != nans:
    new_data.append(row[:])
```

```python
# checking of there is any values after "+" in column TS_or_Distance
culprit = [False, None]
for time_ in data_after_dropping["TS_or_Distance"]:
  two_parts = time_.split("+",)
  if two_parts[1] != "00:00":
    culprit = [True, time_]
if culprit[0]:
  print (culprit[1])
```

```python
def split_a_utc_time(time_):
  preprocessed = time_.split("+",)[0]
  date_time = preprocessed.split(" ",)
  year_, month_, date_ = [int(number) for number in date_time[0].split("-", )]
  hour_, minute_, second_ = [int(number) for number in date_time[1][:date_time[1].index('.')].split(":", )]
  mseconds_ = int(date_time[1].split('.', )[1])
  result = {
      "year": year_,
      "month": month_,
      "date": date_,
      "hour": hour_,
      "minute": minute_,
      "second": second_,
      "mseconds": mseconds_
  }
  return result
```

```python
 # flattening the TS_or_Distance
target_column = "TS_or_Distance"
cells = [
  "year", "month", "date", "hour", "minute", "second", "mseconds"
]
for time_cell_name in cells:
  data_after_dropping[target_column+"_"+time_cell_name] = 0
```

```python
# splitting TS_or_Distance into multiple values and assigning for every rows
for index in range(number_of_rows):
  time_cells = split_a_utc_time(data_after_dropping[target_column][index])
  for time_cell_name in cells:
    data_after_dropping[target_column+"_"+time_cell_name][index] = time_cells[time_cell_name]
```

```python
# dropping TS_or_Distance after splitting it into multiple columns
# we shall drop @ts also as it reflects the same timeframe as TS_or_Distance
data_after_dropping = data_after_dropping.drop(["TS_or_Distance", "@ts"], axis=1)
```
